```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
```

```python
data=pd.read_csv('exit_polls_data_cnn_04022017.csv')
def information_calculator(px):
    return -px*np.log2(px)-(1-px)*np.log2(1-px)

p_x=np.linspace(0,1,10000)
#h_x=(p_x*-np.log2(p_x))*2
q_x=np.abs(1-p_x)
#h_x=-p_x*np.log2(p_x)-q_x*np.log2(q_x)
h_x=information_calculator(p_x)
```

```python
dataslice=data[data.state=='nation']
colorz =["#2c7bb6","#00a6ca","#00ccbc","#90eb9d","#f9d057","#f29e2e", "#e76818","#d7191c","#7C26CB"];
dataslice=dataslice[dataslice.questions=='Race and gender']
dataslice=dataslice.sort_values(by=['Trump_perc'],ascending=False)
fig,(ax,ax2)=plt.subplots(1,2,figsize=(8,3))
values=1-information_calculator(dataslice['Trump_perc']/100)
ax.plot(p_x,1-h_x,color='black',alpha=0.5)
ax.scatter(dataslice['Trump_perc']/100,values,color=colorz,s=80)
ax.set_xlabel('Percentage Voted for Trump')
ax.set_ylabel('Information (Bits)')

ax2.bar(range(len(values)), values, align='center',color=colorz)
plt.xticks(range(len(values)),dataslice.options,rotation='vertical')

fig.savefig('output_color.pdf')
```
